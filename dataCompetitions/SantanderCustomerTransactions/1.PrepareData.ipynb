```python
import pandas as pd 
import numpy as np

from sklearn.model_selection import GridSearchCV

#logloss
from sklearn.metrics import log_loss  

from matplotlib import pyplot
import seaborn as sns
%matplotlib inline
```

```python
# path to where the data lies
dpath = '../../../data/customertransact/'
train = pd.read_csv(dpath +"train.csv")
train.head()
```

```python
train.describe()
```

```python
y_train = train["target"]

train = train.drop(["target"], axis=1).drop(["ID_code"], axis = 1)
X_train = np.array(train)
X_train
```

```python
# 数据标准化
from sklearn.preprocessing import StandardScaler

# 初始化特征的标准化器
ss_X = StandardScaler()

# 分别对训练和测试数据的特征进行标准化处理
X_train = ss_X.fit_transform(X_train)
X_train
```

```python
from joblib import dump

dump(X_train, './X_train.joblib_dat')
dump(y_train, './y_train.joblib_dat')
```

```python
test = pd.read_csv(dpath +"test.csv")
test.head()
```

```python
test = test.drop(["ID_code"], axis = 1)
X_test = np.array(test)
X_test
```

```python
X_test = ss_X.fit_transform(X_test)
X_test
```

```python
dump(X_test, './X_test.joblib_dat')
```
