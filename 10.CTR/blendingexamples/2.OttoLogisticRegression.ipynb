```python
from joblib import dump, load

X_train = load('./X_train.joblib_dat')
y_train = load('./y_train.joblib_dat')
```

```python
from sklearn.linear_model import LogisticRegression

lr_cv = LogisticRegression(penalty='l1', solver='liblinear', multi_class='ovr')
lr_cv.fit(X_train, y_train)
```

```python
lrPredict = lr_cv.predict_proba(X_train[:,:])
```

```python
dump(lrPredict, './lrPredict.joblib_dat')
```
